```python
#train_test_split
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
```

```python
#the model
from sklearn.ensemble import RandomForestRegressor
RF=RandomForestRegressor()
```

```python
#hyperparameters
n_estimators=[int(x) for x in np.linspace(start=100,stop=1200,num=12)]
print(n_estimators)
```

```python
#RandomizedSearchCv applied for hyperparameters tuning
from sklearn.model_selection import RandomizedSearchCV
```

```python
#number of trees in random forest
n_estimators=[int(x) for x in np.linspace(start=100,stop=1200,num=12)]
#number of features to consider at every split
max_features=['auto','sqrt']
#maximum number of levels in trees
max_depth=[int(x) for x in np.linspace(5,3,num=6)]
#minimum number of samples required to split a node
min_samples_split=[2,5,10,15,100]
#minimum number of samples required at each leaf node
min_samples_leaf=[1,2,5,10]
```

```python
#create the random grid
random_grid={'n_estimators':n_estimators,
             'max_features':max_features,
             'max_depth':max_depth,
             'min_samples_split':min_samples_split,
             'min_samples_leaf':min_samples_leaf}
print(random_grid)
```

```python
#use random grid to search for best hyperparameters
#create the base model to tune
rf=RandomForestRegressor()
```

```python
#RandomizedSearchCv applied
rf_random=RandomizedSearchCV(estimator=rf,param_distributions=random_grid,scoring='neg_mean_squared_error',
                             n_iter=15,cv=5,verbose=2,
                            random_state=42,n_jobs=1)
```

```python
rf_random.fit(X_train,Y_train)
```

```python
#prediction
predictions=rf_random.predict(X_test)
predictions
```

```python
#visualize distribution
sns.distplot(Y_test-predictions) # a normal distribution
#the difference between y test and the prediction should be minimal,this what the graph tell us
```
